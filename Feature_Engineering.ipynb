# Log Transformation

##### Log transformation can be used to treat skewed data(outliers)

```python
data = {
    'Value': [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
}
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.DataFrame(data)
```

```python
sns.distplot(df)
```

```python
sns.boxplot(df)
```

```python
#Applying logarithmic transformation
mydata=np.log10(df)
mydata
```

```python
sns.distplot(mydata)
```

```python
sns.boxplot(mydata)
```

# Encoding Categorical Values

## Ordinal Encoder

```python
mydata = {'Education': ['High School', 'Bachelor', 'Master', 'Bachelor']}
df = pd.DataFrame(mydata)
```

```python
education_order=["Master","Bachelor","High School"]
```

```python
df["encoded"]=df["Education"].apply(lambda x: education_order.index(x))
```
